```python
import matplotlib.pyplot as plt
import pandas
import pynumdiff
import seaborn as sns
import numpy as np
```

```python
# colors

# Save a palette to a variable:
color_p = 'Paired'
color_range = 15
palette = sns.color_palette(color_p,color_range)

# Use palplot and pass in the variable:
sns.palplot(palette)

groomer_color = sns.color_palette(color_p,color_range)[1]
icy_color = sns.color_palette(color_p,color_range)[4]
powder_color = sns.color_palette(color_p,color_range)[5]
slushy_color = sns.color_palette(color_p,color_range)[9]
```

```python
fnames = ['DATA/df_grmr.hdf',
          'DATA/df_grmr1.hdf',
          'DATA/df_grmr2.hdf',
          'DATA/df_grmr3.hdf',
          'DATA/df_grmr4.hdf', 
          'DATA/df_grmr5.hdf',
          'DATA/df_icy.hdf',
          'DATA/df_pwdr.hdf', 
          'DATA/df_pwdr1.hdf',
          'DATA/df_s.hdf',
          'DATA/df_s1.hdf', 
          'DATA/df_s2.hdf',
          'DATA/df_s3.hdf']

elevation_fnames = []
for fname in fnames:
    efname = fname[0:-4] + '_speed.hdf'
    elevation_fnames.append(efname)
```

```python
fname_to_type = { 'DATA/df_grmr_speed.hdf': 'groomer slow',
                  'DATA/df_grmr1_speed.hdf': 'groomer slow',
                  'DATA/df_grmr2_speed.hdf': 'groomer fast',
                  'DATA/df_grmr3_speed.hdf': 'groomer fast',
                  'DATA/df_grmr4_speed.hdf': 'groomer slow', 
                  'DATA/df_grmr5_speed.hdf': 'groomer fast',
                  'DATA/df_icy_speed.hdf': 'icy fast',
                  'DATA/df_pwdr_speed.hdf': 'powder slow', 
                  'DATA/df_pwdr1_speed.hdf': 'powder slow',
                  'DATA/df_s_speed.hdf': 'slushy fast',
                  'DATA/df_s1_speed.hdf': 'slushy slow', 
                  'DATA/df_s2_speed.hdf': 'slushy fast',
                  'DATA/df_s3_speed.hdf': 'slushy slow'}
```

```python
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111)

for fname in elevation_fnames:
    try:
        df = pandas.read_hdf(fname)
    except:
        continue
        
    if 'grmr' in fname:
        color = groomer_color
    elif 'icy' in fname:
        color = icy_color
    elif 'pwdr' in fname:
        color = powder_color
    elif '_s' in fname:
        color = slushy_color
        
    ax.plot(df.distance[1000:-1000], df.speed[1000:-1000], color=color)
```

```python
speed_groomer_slow = []
speed_groomer_fast = []
speed_icy = []
speed_powder = []
speed_slushy_slow = []
speed_slushy_fast = []
```

```python


for fname in elevation_fnames:
    try:
        df = pandas.read_hdf(fname)
    except:
        continue
        
    if 'grmr' in fname:
        color = groomer_color
    elif 'icy' in fname:
        color = icy_color
    elif 'pwdr' in fname:
        color = powder_color
    elif '_s' in fname:
        color = slushy_color
        
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111)
    ax.plot(df.lon, df.lat, color=color)
    ax.set_title(np.median(df.speed))
    
    if fname_to_type[fname] == 'groomer slow':
        speed_groomer_slow.append(np.median(df.speed))
    if fname_to_type[fname] == 'groomer fast':
        speed_groomer_fast.append(np.median(df.speed))
    if fname_to_type[fname] == 'icy fast':
        speed_icy.append(np.median(df.speed))
    if fname_to_type[fname] == 'powder slow':
        speed_powder.append(np.median(df.speed))
    if fname_to_type[fname] == 'slushy slow':
        speed_slushy_slow.append(np.median(df.speed))
    if fname_to_type[fname] == 'slushy fast':
        speed_slushy_fast.append(np.median(df.speed))
    
    print(fname, np.median(df.speed))
```
